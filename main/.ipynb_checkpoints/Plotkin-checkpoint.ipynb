```python
def drift_x(x, x2, xN, x2N, params):
    auxR = (1 - params['alpha']*(1 + N/params['M']))*(params['betas'][0] - params['betas'][1])
    auxRx = auxR * (x - x2)
    auxD = (1 + params['alpha']*(1 + N/params['M']))*(params['betas'][0] - params['betas'][1])
    auxDx = auxD * (xN - x2N)
    return auxRx - auxDx
    
def drift_N(N, n1, params):
    aux = (1 - params['alpha']*(1 + N/params['M']))
    R = params['betas'][0] * aux
    Rt = params['betas'][1] * aux
    return n1*R + (N-n1)*Rt

def integrate(dt, obs, init_x, init_N, max_time):
    traj_N, traj_x = [init_N], [init_x]
    av_n = obs[0]
    av_N = obs[0]
    t = 0
    while t < max_time:
        ax = drift_x(traj_x[-1], traj_N[-1], params)
        aN = , drift_N(traj_x[-1], traj_N[-1], params)
        Dx, DN = diff_x(traj_x[-1], traj_N[-1], params), diff_N(traj_x[-1], traj_N[-1], params)
        dvarx = (2*traj_x[-1]*ax + 2*Dx) * dt
        dvarN = (2*traj_N[-1]*aN + 2*DN) * dt
        traj_N.append(traj_N[-1] + aN * dt)
        traj_x.append(traj_x[-1] + ax * dt)
        traj_varN.append(traj_varN[-1] + dvarN)
        traj_varx.append(traj_varx[-1] + dvarx)
        t += dt
    return np.array(traj_x), np.array(traj_N), np.array(traj_varx), np.array(traj_varN)
```

```python
def N_weights(N, a, M, scale):
    aux1 = -2*N + 4*M/a*np.log(1+a+a*N/M)
    aux2 = -np.log(N/M) - np.log(1+a+a*N/M)
    return np.exp((aux1+aux2)-scale)

def N_weights_appr(N, a, M, scale):
    aux1 = -2*N + 4*M/a*np.log(1+a+a*N/M)
    return np.exp(aux1-scale)

def pN_approx(y,a,M):
    aux1 = -2*y*M + 4*M/a*np.log(1+a+a*y)
    auxZ = -2*M*(1/a-a) + np.log(np.pi*M/a)/2
    return np.exp(aux1-auxZ)
```
